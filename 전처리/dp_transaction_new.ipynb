```python
# 년도 datetime으로 타입 변경
transaction['created_at'] = pd.to_datetime(transaction['created_at'])
transaction['shipment_date_limit'] = pd.to_datetime(transaction['shipment_date_limit'])
```

```python
# 배송기간 계산
transaction['shipment_date'] = (transaction['shipment_date_limit'] - transaction['created_at']).dt.days
```

```python
# promo_code 결측치 nopromo로 변경
transaction['promo_code'].fillna('nopromo', inplace=True)
```

# 토탈 어마운트 관련 내용 전처리 작업

```python
transaction_t = transaction.copy()
```

```python
# 'created_at', 'promo_amount', 'shipment_fee', 'total_amount' total_amount 컬럼의 중복값 제거
transaction_t = transaction_t.drop_duplicates(subset=['created_at', 'total_amount', 'customer_id'])
```

```python
# 일회성 고객 분류 'customer_id' 컬럼의 중복값 여부를 확인하고 'customer_type' 컬럼에 분류 결과 추가
transaction_t['customer_type'] = '0'
transaction_t.loc[transaction_t['customer_id'].duplicated(keep=False), 'customer_type'] = '1'

transaction_t.head()
```

```python
payment_status_counts = transaction_t['customer_type'].value_counts()
print(payment_status_counts)
```

```python
transaction_t.info()
```

```python
transaction_t.to_csv("transaction_t.csv", index=False)
```

# 제품가격 총합으로 변경해서 전처리 진행

```python
transaction_n = transaction.copy()
```

```python
transaction_n['total_cost'] = transaction_n['item_price'] * transaction_n['quantity']
```

```python
# 제품가격 사용 불가능한 컬럼 제외
transaction_n.drop(columns=['total_amount'], inplace=True)
```

```python
transaction_n.head()
```

```python
# 일회성 고객 분류 'customer_id' 컬럼의 중복값 여부를 확인하고 'customer_type' 컬럼에 분류 결과 추가
transaction_n['customer_type'] = '0'
transaction_n.loc[transaction_n['customer_id'].duplicated(keep=False), 'customer_type'] = '1'

transaction_n.head()
```

```python
payment_status_counts2 = transaction_n['customer_type'].value_counts()
print(payment_status_counts2)
```

```python
def categorize_price(item_price):
    q1 = np.percentile(transaction_n['item_price'], 25)
    q3 = np.percentile(transaction_n['item_price'], 75)
    
    categories = np.select([item_price <= q1, item_price <= q3], ['저가', '중가'], default='고가')
    return categories

# 'price_category' 컬럼 추가하여 벡터화된 연산으로 그룹화
transaction_n['price_category'] = categorize_price(transaction_n['item_price'])
group_counts = transaction_n['price_category'].value_counts()
```
