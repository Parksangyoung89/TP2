```python
# 섹션으 시작 시간과 끝나는 시간 계산
session_st = click_r.groupby('session_id')['event_time'].min()
session_et = click_r.groupby('session_id')['event_time'].max()

# 섹션 시간 저장
click_r['session_t'] = click_r['session_id'].map(session_et - session_st)

# 전체 평균 체류시간을 계산
ast = click_r['session_t'].mean()

print("평균 체류 :", ast)
```

```python
# session_id와 event_time을 기준으로 정렬
df = click_t.sort_values(['session_id', 'event_time'])

# 이전 row의 event_time을 저장할 변수 초기화
pre_event_time = None

# session_id에 번호를 붙일 변수 초기화
session_num = 1

# session_id가 같은 경우, event_time 차이를 계산하고 번호를 붙임
for i, row in df.iterrows():
    if pre_event_time is not None and row['session_id'] == pre_session_id:
        time_diff = row['event_time'] - pre_event_time
        if time_diff >= timedelta(minutes=30):
            session_num += 1
    df.at[i, 'session_id'] = f"{row['session_id']}_{session_num}"
    pre_event_time = row['event_time']
    pre_session_id = row['session_id']

# 결과 출력
print(df)
```

```python
# 각 세션의 시작 시간과 종료 시간 계산
session_st2 = df.groupby('session_id')['event_time'].min()
session_et2 = df.groupby('session_id')['event_time'].max()

# 세션의 지속 시간 계산 및 exist_time 열 생성
df['session_t'] = df['session_id'].map(session_et2 - session_st2)

# 평균 체류 시간 계산
ast2 = df['session_t'].mean()

print("평균 체류 시간 :", ast2)
```

```python
before_ids = click_r['session_id'].nunique()
unique_session_ids = df['session_id'].nunique()

print("기존 session_id의 개수:", before_ids)
print("새로 정의한 session_id의 개수:", unique_session_ids)
```
